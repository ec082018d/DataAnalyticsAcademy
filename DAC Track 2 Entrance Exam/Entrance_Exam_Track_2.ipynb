```python
print("========================")
print("EDA: Plot Cereal Rating:")
print("========================")
plt.figure(figsize=(9,8))
sns.distplot(df['rating'], color='g', bins=100, hist_kws={'alpha': 0.4})
plt.savefig('EDA_documents/Cereal-Rating-dist.png')
```

```python
print("========================")
print("EDA: Plot other interger columns:")
print("========================")
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.hist(figsize=(16,20), bins=50, xlabelsize=8, ylabelsize=8)
plt.savefig('EDA_documents/Plot_other_interger_columns.png')
```

```python
print("========================")
print("EDA: Correlations:")
print("========================")
df_num_corr = df_num.corr()['rating'][:-1]
features = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There are {} strongly correlated values with rating:\n{}".format(len(features), features))

```

```python
print("========================")
print("EDA: Missing Values:")
print("========================")
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Missing Percent'])
missing_data['Missing Percent'] = missing_data['Missing Percent'].apply(lambda x: x * 100)
missing_data.loc[missing_data['Missing Percent'] > 10][:10]
```

```python
print("========================")
print("Non_Nutritional ranking based on sodium, sugar, calories columns:")
print("========================")


df["Cereal_Non_Nutritional_Total"] = df["sodium"] + df["sugars"] + df["calories"]
df["Cereal_Non_Nutritional_Ranking"] = df["Cereal_Non_Nutritional_Total"].rank(method='dense', ascending=True)

df.head()
```

```python
print("========================")
print("Nutritional Ranking based on fiber, potassium, vitamins columns:")
print("========================")


df["Cereal_Nutritional_Total"] = df["fiber"] + df["potass"] + df["vitamins"]
df["Cereal_Nutritional_Ranking"] = df["Cereal_Nutritional_Total"].rank(method='dense', ascending=False)

df.head()
```

```python
df_non_nutritional_ranking = df[['name','type','mfr', 'cups','rating','Cereal_Non_Nutritional_Total','Cereal_Non_Nutritional_Ranking']]
df.sort_values(by=['Cereal_Non_Nutritional_Ranking'], inplace=True, ascending=True)
df_non_nutritional_ranking.head()

```

```python
df_nutritional_ranking = df[['name','type','mfr', 'cups','rating', 'Cereal_Nutritional_Total', 'Cereal_Nutritional_Ranking']]
df.sort_values(by=['Cereal_Nutritional_Ranking'], inplace=True, ascending=True)
df_nutritional_ranking.head()
```

```python
print("========================")
print("Nutritional Ranking Comparison")
print("========================")


df_nutritional_compare_ranking = df[['name','rating','Cereal_Non_Nutritional_Total','Cereal_Nutritional_Total','Cereal_Non_Nutritional_Ranking','Cereal_Nutritional_Ranking']]
df_nutritional_compare_ranking.head()
```

```python
df.to_excel("processed_data/df_output.xlsx")
df_non_nutritional_ranking.to_excel("processed_data/df_non_nutritional_ranking_output.xlsx")  
df_nutritional_ranking.to_excel("processed_data/df_nutritional_ranking_output.xlsx")
df_nutritional_compare_ranking.to_excel("processed_data/df_nutritional_compare_ranking_output.xlsx")
```
